# Single and multiple string

```python
name1 = 'python'
name2 = "python"
name3 = '''python''' # multiple string
name4 = """python""" # multiple string
print(name4)
```

```python
sen = "this is line1."\
"this is line2."\
"this is line3."
print(sen)
```

```python
sen = """this is line1.
this is line2.
this is line3."""
print(sen)
```

# Postive and Negative indexing

```python
word="python"
print(word[0]) # positive indexing
print(word[1]) # positive indexing
print(word[2]) # positive indexing
print(word[3]) # positive indexing
print(word[4]) # positive indexing
print(word[5]) # positive indexing
```

```python
word="python"
print(word[-1]) # negative indexing
print(word[-2]) # negative indexing
print(word[-3]) # negative indexing
print(word[-4]) # negative indexing
print(word[-5]) # negative indexing
print(word[-6]) # negative indexing
```

# Slicing

```python
word = "Python"
# print(word[start:stop:step])
# start default value =0
# stop default value = len(word)
#step default value = 1
print(word[0:3])
print(word[0:5:2])
print(word[1:6:1])
```

```python
word = "Python"
print(word[-1::])
print(word[-3::1])
print(word[-3:-1:1])
```

# String Formatting

```python
name= "shagun chaudhary"
city = "noida"
print("hi name your city is city")
print(f"hi {name} your city is {city}")
```

```python
print("hi {} your city is {}".format(name,city))
```
